```python
import scipy.io.wavfile

import matplotlib.pyplot as plt
#%matplotlib inline
%matplotlib notebook
#from pylab import rcParams
#rcParams['figure.figsize'] = 10, 5
```

```python
rate, whistle_data = scipy.io.wavfile.read('whistle.wav')
print(rate)
```

# FFT

```python
import scipy.fftpack as fftpk
import numpy as np

Z = fftpk.fft(whistle_data[:,0])
magZ = np.absolute(Z)
phsZ = np.angle(Z)

plt.figure()
plt.subplot()
```

# Spectrogram

```python
import scipy.signal as sig

f, t, Sxx = sig.spectrogram(whistle_data[:,1], rate)

plt.figure()
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
