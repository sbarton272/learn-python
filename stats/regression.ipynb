## 3.1 Simple Linear Regression

Good resource: https://www.stat.purdue.edu/~snandy/stat512/topic3.pdf

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data
N = 100
noise = np.random.normal(scale=10, size=(N))
x = np.random.normal(size=(N))
b1 = 8
b0 = 3
y = b1 * x + b0 + noise

x_ax = np.linspace(np.min(x), np.max(x))

plt.scatter(x, y, alpha=0.3, linewidths=0)
plt.plot(x_ax, b1 * x_ax + b0, color='r', label='real')

# Regression
X = np.concatenate([np.ones((N, 1)), x.reshape(-1, 1)], axis=1)
Y = y.reshape(-1, 1)
B = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
plt.plot(x_ax, B[1, 0] * x_ax + B[0, 0], color='g', label='fit')

plt.title('Regression with random samples')
plt.legend()
```
